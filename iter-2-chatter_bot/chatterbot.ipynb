```python
# Reading in the data
df = pd.read_csv('../data/clean_scripts.csv', index_col=0)

# Dropping NaNs
df.dropna(inplace=True)
```

```python
# Examining the data
df.head()
```

```python
# Creating a dialogue length column
df['dialogue_len'] = df['Dialogue'].apply(lambda x: len(x))
```

```python
# Creating a word count column
df['wordcount'] = df['clean_dialogue'].apply(lambda x: len(word_tokenize(x)))
```

```python
# Creating a dataframe with of lines of dialogue less than 50 words long.
less_50  = df[df['wordcount']<50]
```

```python
# Creating a list of clean dialogue
dialogue_list = list(less_50['clean_dialogue'])
```

```python
def seq_prep(dialogue):
    '''
    This function prepares a list of dialogue in the way 
    '''
    dialogue_combos = []
    for i in range(len(dialogue)+ 1):
        try:
            dialogue_combos.append([dialogue[i], dialogue[i+1]])
        except:
            pass
    return dialogue_combos
        
    
```

```python
combos = seq_prep(dialogue_list)
```
