```python
import pandas as pd
import numpy as np
import re
import pickle

from numpy.random import choice

from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('scripts.csv', index_col=0)
```

```python
def string_len(string):
    try:
        return len(string)
    except:
        pass
```

```python
df['dialogue_len'] = df['Dialogue'].apply(string_len)
```

```python
df.dropna(inplace=True)
```

```python
# removing parenthesis
df['removed_paren_dialogue'] = df['Dialogue'].str.replace(r"\(.*\)","")
```

```python
df['wordcount'] = df['removed_paren_dialogue'].apply(lambda x: len(word_tokenize(x)))
```

```python
df = df[df['wordcount']< 16]
```

```python
dialogue_list = list(df['removed_paren_dialogue'])
```

```python
dialogue_combos = []
```

```python
def make_combos(dialogue):
    dialogue_combos = []
    for i in range(len(dialogue)+ 1):
        try:
            dialogue_combos.append([dialogue[i], dialogue[i+1]])
        except:
            pass
    return dialogue_combos
        
    
```

```python
combos = make_combos(dialogue_list)
```

```python
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
from chatterbot.trainers import ListTrainer
```

```python
bot = ChatBot('Seinfeld')
trainer = ListTrainer(bot)
```

```python
choice(3412341)
```

```python
def random_seinfeld(n):
    result = []
    for i in range(n):
        result.append(combos[choice(len(combos))])
    return result
        
```

```python
random_seinfeld(3)
```

```python
with open('combos.pkl', 'wb') as f:
    pickle.dump(combos, f)
```

```python
for combo in combos:
    trainer.train(combo)
```
